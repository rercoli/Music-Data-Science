```python
# create list of dataframes by year
dfs = [df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019]
dfs_count=[]

# create list of dataframes by year with count method applied
for df in dfs:
    dfs_count.append(df.groupby(["top genre"]).count())

# reset index for each dataframe
for i in range(10):
    dfs_count[i] = dfs_count[i].reset_index()   
```

```python
# plot most popular genres by year
for i in range(10):
    fig, ax1 = plt.subplots(figsize=(15,8 ))
    ax1.set_title("top genres in 201" + str(i), fontsize=15, pad=20)
    rects1 = ax1.bar(dfs_count[i]['top genre'], dfs_count[i]['title'], label="201" + str(i))
    ax1.set_ylabel('occurrences in the top 30s')
    ax1.set_xlabel('genres')
    ax1.tick_params(rotation=90)
    ax1.grid()
    plt.tight_layout()

    # function to attach a text label above each bar, displaying its height
    def autolabel(rects):
        #Attach a text label above each bar in *rects*, displaying its height
        for rect in rects:
            height = rect.get_height()
            ax1.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    # call the function
    autolabel(rects1)

    # save figure
    plt.tight_layout()
    plt.savefig(str(i)+".png")

```
